```python
import re
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')

pd.set_option('display.max_rows', None)
```

```python
chat = pd.read_table('amigos_de_inf.txt', header=None)
chat.columns = ['msg']
print(chat.shape)
```

```python
lista = []

date_r = r'\d+\/\d+\/\d+'
time_r = r'\d+\:\d+'
name_r = r'\d{2} - (.*?): '
msg_r = r': (.*)'

msgs = [i for i in chat.msg]

for i in msgs:
    date = re.findall(date_r, i)
    time = re.findall(time_r, i)
    name = re.findall(name_r, i)
    msg = re.findall(msg_r, i)
    if date and time and name and msg:
        lista.append([date[0], time[0], name[0], msg[0]])

wpp = pd.DataFrame(lista, columns=['date', 'time', 'name', 'msg'])
```

```python
wpp['day'] = [x.split("/")[0] for x in wpp.date]
wpp['month'] = [x.split("/")[1] for x in wpp.date]
wpp['year'] = [x.split("/")[2] for x in wpp.date]
wpp['week_day'] = [pd.to_datetime(x).weekday() for x in wpp.date]
wpp['msg_length'] = [len(i) for i in wpp.msg]
wpp['media'] = [('Arquivo de mídia oculto' in i)*1 for i in wpp.msg]
wpp['no_media'] = [(not (('Arquivo de mídia oculto' in i)*1))*1 for i in wpp.msg]
```

```python
nomes = {'Renato Aranha':'Renato', 'Zidane - São Crist': 'Zidane', 'Felippe - Sao Crist': 'Felippe', 'R Massa':'Rafael', '+55 21 96421-4891':'Kayo'}
wpp.name.replace(to_replace=nomes, inplace=True)
```

```python
pd.DataFrame(wpp.groupby(['year','name'])[['media']].sum().sort_values(['year', 'media'], ascending=[True,False]))
```

```python
pd.DataFrame(wpp.groupby('year').name.value_counts())
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16, 5))

for j,i in enumerate(range(2019,2021)):
    df = pd.DataFrame(wpp[wpp['year']==str(i)].name.value_counts())
    qtd_msgs = wpp[wpp['year']==str(i)].msg.count()
    df.plot.bar(fontsize=12, ax=axes[j], title=str(i)+': '+str(qtd_msgs)+' mensagens', colormap='plasma', alpha=0.9)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(12, 6), sharex=True)

for j,i in enumerate(range(2019,2021)):
    i = str(i)
    col = wpp[(wpp.year==i)].time.value_counts()
    col.index = pd.to_datetime(col.index)
    menor = pd.DataFrame(col.resample('30min').sum())
    menor.index = [x.time().strftime('%H:%M') for x in menor.index]
    qtd_msgs = wpp[wpp['year']==i].msg.count()
    menor.plot.bar(fontsize=10, ax=axes[j], title = str(i)+': '+str(qtd_msgs)+' msgs', colormap='plasma', alpha=0.9)
    plt.tight_layout()
```

```python
ano = '2020'
filtro = wpp[(wpp['year']==ano)].name.head(10).unique()
nrows = len(filtro)

fig, axes = plt.subplots(nrows=nrows, ncols=1, figsize=(12, 10))

for i,j in enumerate(filtro):
    col = wpp[(wpp['name']==j)&(wpp['year']==ano)].time.value_counts()
    col.index = pd.to_datetime(col.index)
    col = col.sort_index()
    menor = pd.DataFrame(col.resample('30min').sum())
    menor.index = [x.time().strftime('%H:%M') for x in menor.index]
    qtd_msgs = wpp[(wpp['name']==j)&(wpp['year']==ano)].msg.count()
    menor.plot.bar(fontsize=8, ax=axes[i], title= str(j)+': '+str(qtd_msgs)+' msgs', colormap='plasma', alpha=0.8)
    #axes[i].set_ylabel(j)
    plt.tight_layout()
    plt.xticks(rotation=90)
```
